```python
player_1_win=[("rock","scissor"),("paper","rock"),("scissor","paper")]
from getpass import getpass
player_1_choice=getpass("Enter your choice : ").strip().lower()
player_2_choice=getpass("Enter your choice : ").strip().lower()
choice=(player_1_choice,player_2_choice)
if player_1_choice==player_2_choice:
    print("Match Tie")
elif choice in player_1_win:
    print("Player one won the game")
else:
    print("Player two won the game")
print("The choice of player one was",player_1_choice)
print("The choiceof player two was",player_2_choice)
```

```python
even=[]
counter=15
while counter<=100:
    if counter%2==0:
        even.append(counter)
        counter=counter+1
    else:
        counter+=1
print(even)    
```

```python
odd=[]
counter=15
while counter<=100:
    if counter%2==0:
        counter+=1
    else:
        odd.append(counter)
        counter=counter+1
print(odd)    
```
