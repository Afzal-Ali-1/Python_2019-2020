```python
transaction_limit=int(input("Enter your number where you want to stop execution or addition"))
counter=0
transaction_count=0
while counter<transaction_limit:
    transaction_amount=int(input("Enter the number : "))
    transaction_count=transaction_count+transaction_amount
    counter=transaction_count
print(transaction_count)
```
