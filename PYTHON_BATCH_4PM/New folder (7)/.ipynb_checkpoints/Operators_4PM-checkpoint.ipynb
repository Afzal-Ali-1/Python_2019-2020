```python
number=int(input("Enter the number : "))
if number>1:
    for var in range(2,number):
        if number%var==0:
            print("Given number is not prime")
            break
    else:
        print("Given number is a prime number")
```

```python
number=int(input("Enter the number : "))
if number>1:
    for var in range(2,(number//2+1)):
        if number%var==0:
            print("Given number is not prime")
            break
    else:
        print("Given number is a prime number")
```
